Predict song popularity

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

Due to the data set having more than 50mb the link to download it is in the file and in the Read me

```python
data = pd.read_csv('spotify_data.csv')
data.shape
data.info()
data.describe()
```

```python
data.columns
data.head(5)
```

```python
data = data.drop(columns=['track_id', 'artist_name', 'track_name'])
```

Preparing data for Model

```python

codificador = LabelEncoder()
data['genre'] = codificador.fit_transform(data['genre'])
data.info()
```

```python
y = data['popularity']
x = data.drop(['popularity', 'Unnamed: 0'], axis=1)

x_training, x_test, y_training, y_test = train_test_split(x, y, test_size = 0.4)
```

```python
#Arvore de decis√£o
decision_tree = RandomForestClassifier()
#KNN
knn_model = KNeighborsClassifier()

decision_tree.fit(x_training, y_training)
knn_model.fit(x_training, y_training)
```

```python
decision_tree_prediction = decision_tree.predict(x_test)
knn_model_prediction = knn_model.predict(x_test.to_numpy())
print(f'deciciosn tree acc is: ', accuracy_score(y_test, decision_tree_prediction))
print(f'knn model acc is: ', accuracy_score(y_test, knn_model_prediction))
```

Utilizando esses dois modelos pode-se observar que estes n√£o possuem uma acur√°cia estat√≠stica confiavel, ficando com 15,89% do modelo de arvore de decis√£o e aproximadamente 9% no modelo KMM
